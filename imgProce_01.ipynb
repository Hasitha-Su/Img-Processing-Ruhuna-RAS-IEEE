```
import numpy as np
```

```
img=cv2.imread("colour.png",1)
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV) #convert image to hsv model

#define range lower and upper
#detect red only 
# lower_red = np.array([119,27,45])
# upper_red = np.array([208,255,255])
# mask = cv2.inRange(hsv,lower_red,upper_red) #create a mask 

#detect green only
lower_green = np.array([25,52,72])
upper_green = np.array([102,255,255])
mask = cv2.inRange(hsv,lower_green,upper_green)

#bitwise operation on the mask to get a colour image
img2 = cv2.bitwise_and(img, img, mask=mask)

cv2_imshow(img)
cv2_imshow(img2)
```
